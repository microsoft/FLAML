# Visualization with FLAML
The `flaml.visualization` module provides utility functions for plotting the optimization process using [plotly](https://plotly.com/python/).  Leveraging `plotly`, users can interactively explore experiment results. To use these plotting functions, simply provide your Hyperparameter Tuning & AutoML experiment results as input. Optional parameters can be added using keyword arguments.

## Avaliable Plots
1. [Optimization History](#Optimization-History): Plot optimization history of all trials in the experiment.
2. [Feature Importance](#Feature-Importance): Plot importance for each feature in the dataset.
3. [Parallel Coordinate](#Parallel-Coordinate): Plot the high-dimensional parameter relationships in the experiment.
4. [Contour](#Contour): Plot the parameter relationship as contour plot in the experiment.
5. [EDF](#EDF): Plot the objective value EDF (empirical distribution function) of the experiment.
6. [Timeline](#Timeline): Plot the timeline of the experiment.
7. [Slice](#Slice): Plot the parameter relationship as slice plot in a study.
8. [Hyperparameter Importance](#Hyperparameter-Importance): Plot the hyperparameter importance of the experiment. This plot use f-ANOVA to evaluate hyperparameter importance. 

## Conduct AutoML Experiment 

```python
# # uncomment and try different renderers if the figure is not shown in the next few cells
# import plotly.io as pio
# pio.renderers.default = "notebook"  # "notebook", "sphinx_gallery"
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import warnings
from flaml import AutoML

warnings.filterwarnings("ignore")
x, y = load_iris(return_X_y=True, as_frame=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7654321)

aml = AutoML()
automl_settings = {
    "time_budget": 10,
    "task": "classification"
}
aml.fit(X_train=x_train, y_train=y_train, **automl_settings)
```

## Visualize Experiment Result

```python
import flaml.visualization as fviz
```

### Optimization History
Plot optimization history of all trials in the experiment.

```python
fig = fviz.plot_optimization_history(aml)
# or
fig = fviz.plot(aml, "optimization_history")
fig.show()
```

### Feature Importance
Plot importance for each feature in the dataset.

```python
fig = fviz.plot_feature_importance(aml)
# or
fig = fviz.plot(aml, "feature_importance")
fig.show()
```

### Parallel Coordinate
Plot the high-dimensional parameter relationships in the experiment.


Extra argument for this function:
- `learner`: Specify the learner you intend to study in the experiment. Only applicable in AutoML experiment result. Leaving this blank will choose the best learner in the whole experiment by default.
- `params`: A list to specify which hyperparameter to display. Leaving this blank will display all hyperparameters avaliable.



```python
fig = fviz.plot_parallel_coordinate(aml, learner="lgbm", params=["n_estimators", "num_leaves", "learning_rate"])
# or
fig = fviz.plot(aml, "parallel_coordinate", learner="lgbm", params=["n_estimators", "num_leaves", "learning_rate"])
fig.show()
```

## Contour
Plot the parameter relationship as contour plot in the experiment.



Extra argument for this function:
- `learner`: Specify the learner you intend to study in the experiment. Only applicable in AutoML experiment result. Leaving this blank will choose the best learner in the whole experiment by default.
- `params`: A list to specify which hyperparameter to display. Leaving this blank will display all hyperparameters avaliable.

```python
fig = fviz.plot_contour(aml, learner="lgbm", params=["n_estimators", "num_leaves", "learning_rate"])
# or
fig = fviz.plot(aml, "contour", learner="lgbm", params=["n_estimators", "num_leaves", "learning_rate"])
fig.show()
```

### Empirical Distribution Function
Plot the objective value EDF (empirical distribution function) of the experiment.
EDF is useful to analyze and improve search spaces. For instance, you can see a practical use case of EDF in the paper [Designing Network Design Spaces](https://arxiv.org/abs/2003.13678).


- For AutoML experiment, trials of each learner is an optimazation series.
- For hyperparameter tuning, you can pass one or many(in list or dictionary) to this function.

```python
fig = fviz.plot_edf(aml)
# or
fig = fviz.plot(aml, "edf")
fig.show()
```

### Timeline
Plot the timeline of the experiment.

```python
fig = fviz.plot_timeline(aml)
# or
fig = fviz.plot(aml, "timeline")
fig.show()
```
